````python
# URL, TEXT, QUESTIONS: [QUESTION, ANSWER]
collection = []
for data in js['data']:
    topic = data['title']
    for para_no, para in enumerate(data['paragraphs'], start=1):
        raw_text = para['context']
        
        doc = {
            'TOPIC': topic,
            'ID': para_no, 
            'TEXT': raw_text, 
            'QUESTIONS': []
        }
        for qas_no, qas in enumerate(para['qas'], start=1):
            raw_question = qas['question']
            
            # answers always contains single element inside the list
            raw_answer = qas['answers'][0]['text']
            
            qas_doc = {
                'ID': qas_no,
                'QUESTION': raw_question,
                'ANSWER': raw_answer
            }
            doc['QUESTIONS'].append(qas_doc)
            
        collection.append(doc)
````
